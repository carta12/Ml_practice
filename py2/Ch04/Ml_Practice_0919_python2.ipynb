# 4.나이브베이스: 확률이론으로 분류하기

## 4.1 베이지안 의사결정 이론으로 분류
+ 장점 : 소량의 데이터를 가지고 작업이 이루어지며, 여러 개의 분류항목을 다룰 수 있음
+ 단점 : 입력데이터에 따라 민감하게 작용

##  4.2 조건부 확률
+ 조건부 확률은 어떤 사건 B가 일어났을 때 사건 A가 일어날 확률을 의미한다. 사건 B가 발생했을 때 사건 A가 발생할 확률은 사건 B의 영향을 받아 변하는데 이를 조건부 확률이라 한다.


+ 확률 공간 Ω에서의 두 사건 A, B에 대해서 {\displaystyle P(B)>0} P(B)>0일 때 사건 B가 일어났을 때 사건 A의 조건부 확률은 다음과 같이 정의 된다

## 4.4 나이브 베이스로 문서 분류하기
+ 이메일 구분
    1. 수집 : RSS자료 사용
    2. 준비 : 명목형 or Boolean값
    3. 분석 : 히스토그램으로 확인
    4. 훈련 : 각 속성을 톡립적으로 조건부 확률 계산
    5. 검사 : 오류율 계산
    6. 사용 : 문서구분에 활용

## 4.5 파이썬을 활용한 텍스트 분류
+ 하나의 단어를 하나의 토큰(token)으로 분류
+ 단어가 아닌(자체의 의미가 없는 ex|URL, IP주소 등) 것들도  토큰으로 분류
+ 문서에 존재하는 단어는 1, 그렇지 않은 것은 0으로 구분
